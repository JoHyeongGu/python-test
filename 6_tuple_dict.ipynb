```python
my_tuple = ("AAA", "BBB", "CCC")
print(type(my_tuple))
print(my_tuple, end = "\n\n")

my_tuple = "AAA", "BBB", "CCC"
print(type(my_tuple))
print(my_tuple, end = "\n\n")

my_tuple = tuple("A")
print(type(my_tuple))
print(my_tuple, end = "\n\n")

my_tuple = (10)
print(type(my_tuple))
print(my_tuple, end = "\n\n")

my_tuple = (10,)
print(type(my_tuple))
print(my_tuple, end = "\n\n")

my_tuple = (10, ("Hello", "World"))
print(type(my_tuple))
print(my_tuple)
print(my_tuple[1][0])
```

```python
my_tuple = (100, 200, 300, 400 ,500)
print(my_tuple[1:3])
print(my_tuple[:3])
print(my_tuple[1:])
print(my_tuple[:])
print(my_tuple[-5: -3])
```

```python
tuple_a = (10, 20, 30, 40, 50)
print(tuple_a)
# my_tuple.append(60) => Error
# my_tuple[5] = 60 => Error
# del(my_tuple[0]) => Error
tuple_b = tuple_a
del(tuple_a)
print(tuple_b)
# print(tuple_a) => 삭제해서 출력 불가
tuple_a = ("A", "B")
tuple_c = tuple_a + tuple_b
print(tuple_c)
tuple_c = tuple_a * 3
print(tuple_c)
```

```python
from random import randint

my_list = [randint(1, 100) for e in range(10)]
my_tuple = tuple(my_list)
print("튜플:", my_tuple)
print("정렬된 튜플:", sorted(my_tuple))
print("합:", sum(my_tuple))
print("최대:", max(my_tuple))
print("최소:", min(my_tuple))
print("평균:", sum(my_tuple) / len(my_tuple))
```

```python
my_dict = dict()
print(my_dict)

my_dict["name"] = "JHG"
my_dict["phone"] = "010-0000-0000"
my_dict["birth"] = "0000-00-00"
print(my_dict)

print("name" in my_dict)
print(my_dict["name"])
```

```python
print(my_dict)
my_dict[2] = "b"
my_dict["birth"] = "aaa"
print(my_dict)
```

```python
dic = {"phone" : "000000000", "birth": "111111"}
dic["name"] = "JHGG"
del dic["phone"]
dic[2] = "b"
print(dic)
dic.clear()
print(dic)
del dic
# print(dic) => Error
```

```python
dic = {"phone" : "000000000", "birth": "111111"}
print(dic.keys())

# 반복문으로 dict 타입 순회하기
for (k, v) in dic.items():
  print(k)
  print(v)

# 반복문으로 dict key
for key in dic.keys():
  print(key)

# 반복문으로 dict value
for value in dic.values():
  print(value)
```
