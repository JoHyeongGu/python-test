```python
# 저번 시간 복습하기
info = {
    "name": "홍길동",
    "age": 21,
    "major": "컴퓨터공학",
    "grades": {
      "math": 95,
      "english": 88,
      "science": 92,
    },
}
print("이름:", info["name"])
print("전공:", info["major"])
info["gpa"] = 3.8
print("GPA 추가 후:", info)
info["age"] = 22
print("나이 수정 후:", info)
del info["grades"]
print("성적 삭제 후:", info)
print(*[f"{k} : {v}" + "\n" for k, v in info.items()], sep = "")
```

```python
# Ex01. 합격/불합격 판단하기
score = int(input("정수를 입력하세요 : "))
print(f"{score}점은(는) 합격입니다.\n축하합니다." if score >= 60 else f"{score}점은(는) 불합격입니다.")
```

```python
print("1. Espresso \n2. Americano \n3. Cappuccino")

num = int(input("번호 입력 : "))
menu = ""
if num == 1:
  menu = "Espresso"
elif num == 2:
  menu = "Americano"
elif num == 3:
  menu = "Cappuccino"

print(f"선택한 메뉴는 {menu}" if menu != "" else "나가주세요.")
```

```python
from datetime import date
today = date.today()
year = int(input("몇 년도에 태어났나요? "))
age = today.year - year + 1
if (age >= 15 and age < 20): print("당신은 청소년입니다.")
print(f"올해 당신은 {age}살입니다.")
```

```python
print("1. Espresso \n2. Americano \n3. Cappuccino")

while True:
  num = int(input("번호 입력 : "))
  menu = ""
  if num == 1:
    menu = "Espresso"
  elif num == 2:
    menu = "Americano"
  elif num == 3:
    menu = "Cappuccino"
  if menu != "":
    print("선택한 메뉴는", menu)
  else:
    print("나가주세요.")
    break;
```
