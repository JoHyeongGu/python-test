```python
# 저축 후 원리금 계산
"""
원금 = 500만원
이자 = 원금 * 0.0325
원리금 = 원금 + 이자
원리금이 얼마인가?

출력
원금은 **
원리금은 **
"""

ori = int(input("원금을 입력하세요:"))
print(f"원금은 {ori}")
ori += ori * 0.0325
print(f"원리금은 {ori}")
```
